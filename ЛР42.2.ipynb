#### 1. Дайте определение дифракции. 

Дифракцией называется огибание волнами препятствий. В более широком смысле - любые отклонения при распространении волн от законов геометрической оптики при распространении света в среде с резкими неоднородностями.

#### 2. Что такое волновая поверхность, фронт волны, луч. 

- Волновая поверхность – поверхность, на которой лежат точки, колеблющиеся в одной фазе.
- Фронт волны - это поверхность, до которой дошли колебания к данному моменту времени.
- Луч - это линия, вдоль которой распространяется поток энергии, испущенной источников света или звука. 

#### 3. Какие волны называются плоскими, сферическими?

 Плоская волна - это волна, у которой направление распространения одинаково во всех точках пространства
 Сферическая волна - волна, фронт которой представляет собой сферу.

#### 4. Сформулируйте принцип Гюйгенса. Чему равна амплитуда вторичной сферической волны. 

Принцип Гюйгенса: каждая точка, до которой доходит волна, является точечным источником вторичных волн, а огибающая этих волн дает положение волнового фронта в следующий момент времени.

Амплитуда вторичных волн пропорциональна амплитуде первичной волны $а_0$, приходящей в точку $А$, площади элемента $dS$, убывает с возрастанием угла $\varphi$ между нормалью к поверхности $S$ и направлением излучения вторичной волны на точку $Р$.

#### 5. Чем отличается дифракция Френеля от дифракции Фраунгофера? Какой тип дифракции наблюдается в работе? 

Если рассматривается дифракция сферических или плоских волн и при этом точка наблюдения находится на конечном расстоянии от преграды, то явления такого рода называют дифракцией Френеля. 

Если источник света и точка наблюдения расположены от преграды настолько далеко, что лучи, падающие на преграду, и лучи, идущие в точку наблюдения, образуют практически параллельные пучки, то говорят о дифракции Фраунгофера.
В работе наблюдается дифракция Фраунгофера.

#### 6. Как располагаются спектры для различных длин волн относительно друг друга?

Спектры для коротких длин волн (например, синего света) будут находиться ближе к центральной оси (нулевому порядку), в то время как для длинных длин волн (например, красного света) максимумы будут находиться дальше от центра. 

Если на решетку падает белый свет, то все максимумы, кроме центрального, разлагаются в спектр и пространственно разделены.

#### 7. Дайте определение дифракционной решетки, периода дифракционной решетки. Как связан период дифракционной решетки с числом штрихов на единицу длины?

Дифракционная решетка – спектральный прибор, предназначенный для разложения света в спектр и измерения длин волн. Она представляет собой множество одинаковых равноотстоящих друг от друга прозрачных щелей в непрозрачном экране.

Период дифракционной решетки — это сумма ширины прозрачного участка и ширины штриха.

Величина периода дифракционной решётки связана с числом нанесённых на неё штрихов соотношением $d={1 \over N}$ , где $N$ - число штрихов, приходящихся на единицу длины решетки.

#### 8. Как получаются максимумы интенсивности от дифракционной решетки или нескольких щелей? 

Максимумы интенсивности от дифракционной решётки или нескольких щелей получаются благодаря интерференции световых колебаний в пучках, приходящих в фокальную плоскость линзы от отдельных щелей.

#### 9. Как получаются минимумы и вторичные максимумы от дифракционной решетки? Под какими углами они возникают?

Минимумы и вторичные максимумы от дифракционной решётки получаются в результате интерференции вторичных волн, испускаемых различными точками решётки. В зависимости от разности хода между ними волны усиливают или ослабевают друг друга. 

Интенсивность дополнительных максимумов значительно ниже интенсивности главных, вследствие чего дополнительные максимумы на экране обычно не наблюдаются, а создают весьма слабый фон.

Углы, под которыми наблюдаются главные минимумы,  удовлетворяют условию $dsin\phi=\pm m+ {1 \over 2}\lambda$ .

Углы, под которыми наблюдаются второстепенные минимумы, удовлетворяют условию $dsin\phi=\pm m+ {p \over N}\lambda$ ,$p=1,2,...,N-1$ , где $N$ - число штрихов решётки. 

#### 10. Сколько второстепенных минимумов располагается между двумя главными максимумами в спектре дифракционной решетки? От чего зависит интенсивность вторичных максимумов?

Между двумя главными максимумами в спектре дифракционной решётки располагается (N − 1) второстепенных минимумов, разделенных вторичными максимумами. 
С увеличением числа щелей в решетке интенсивность вторичных максимумов убывает.

## Предпроцессинг данных и подключение библиотек

Код подключает необходимые библиотеки и проводит обработку данных из таблиц с данными, полученными при эксперименте.

```python
from csv import *
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def file_read(path):
    with open(path, 'r') as file:
        data = file.readlines()
    data[0] = data[0][3:]
    #print(data)
    #data = data[1:]
    for i in range(len(data)):
        data[i] = data[i].replace(',','.')
        data[i] = data[i].split(';')
        for j in range(len(data[i])):
            data[i][j] = float(data[i][j])
    #print(data)
    return data
```
